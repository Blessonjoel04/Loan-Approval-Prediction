# Conclusion

```python
models = [log, knn, svc, decision_tree, rf, gbc, xgb]
model_names = ["LogisticRegression", "KNeighborsClassifier", "SVC", "DecisionTreeClassifier", 
               "RandomForestClassifier", "GradientBoostingClassifier", "XGBClassifier"]
results_dict = {}

for model, name in zip(models, model_names):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Get classification report as a dictionary
    report = classification_report(y_test, y_pred, output_dict=True)
    
    # Extract relevant metrics and store in the results dictionary
    results_dict[name] = {
        "Accuracy": report["accuracy"],
        "Class 1 Recall": report["1"]["recall"],
        "Class 1 F1-Score": report["1"]["f1-score"],
        "Weighted F1-Score": report["weighted avg"]["f1-score"]
    }

# Convert the results dictionary to a pandas DataFrame
results_df = pd.DataFrame(results_dict).T
results_df
```

```python
results_df
```

```python
plt.figure(figsize=(20, 6))

# Loop through models to generate confusion matrix and heatmap
for i, model in enumerate(models):
    # Predict with the model
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Generate confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Plot heatmap
    plt.subplot(2, 4, i+1)  # Create subplots, 2 rows and 4 columns
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, xticklabels=["Declined", "Approved"], yticklabels=["Declined", "Approved"])
    plt.title(f'Confusion Matrix for {model_names[i]}')
    plt.xlabel('Predicted')
    plt.ylabel('True')

# Adjust the layout to fit all subplots
plt.tight_layout()
plt.show()
```

1. As per the metrics, `XGBoost` has the highest accuracy. Accuracy indicates the overall performance of the model. But we can't solely rely on the accuracy itself when working with classification problem. F1 score and Recall are generally considered for classification problems. 

2. Class 1 Recall (Sensitivity): This is crucial when predicting loan approvals. A high Class 1 Recall ensures that most of the loans that should be approved are correctly identified, minimizing the number of false negatives (i.e., approved loans being declined).

3. Class 1 F1-Score: The F1-Score balances Precision and Recall. If you want to make sure the model not only recalls most approved loans but also has a good balance between correctly identifying them and avoiding false positives, this metric is important.

**RandomForestClassifier and XGBClassifier slightly outperforms all the other models. Hence the optimal choice would be either `Random Forest` or `XGBoost`**
